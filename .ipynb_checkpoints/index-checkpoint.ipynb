```python
import pandas as pd
import calendar
import seaborn as sns
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
```

## Movie Budgets

```python
data = pd.read_csv(r'zippedData\tn.movie_budgets.csv.gz')
print(data)
```

```python
df_budget = pd.DataFrame(data)
df_budget.head()
```

```python
df_budget.info()

#There is no missing data, or na values
```

```python
df_budget.shape
df_budget.value_counts()
```

```python
# Check to make sure the movie field doesn't contain blanks and NaNs
pd.isna(df_budget['movie'])
```

```python
# We can drop the Worldwide Domestic Gross column since we're focused in the Worldwide Gross
df_budget.drop(['domestic_gross'], axis=1)
```

```python
# Check that none of the movies have a production budget of 0 as they're most likely bad data
df_budget.loc[df_budget['production_budget'] == 0]

# Check that none of the movies have a Worldwide gross of as they're most likely bad data, 

df_budget.loc[df_budget['worldwide_gross'] == 0]
```

```python
df_budget.isna().value_counts()
```

```python
#We want to observe what type of data is the date column so we can work with it
#We want to analyze the date, in order to get the month and year

df_budget['Year'] = pd.DatetimeIndex(df_budget['release_date']).year
df_budget
```

```python
df_budget['Year'].unique()
```

```python
#Converting the month value into string

df_budget['Month'] = pd.DatetimeIndex(df_budget['release_date']).month
df_budget['Month'] = df_budget['Month'].apply(lambda x: calendar.month_abbr[x])
df_budget

```

```python
df_budget['Day'] = pd.DatetimeIndex(df_budget['release_date']).day
df_budget
```

```python
#we clean the worldwide_column and change to float type
df_budget['worldwide_gross'] = df_budget.worldwide_gross.apply(lambda x:x.replace('$','').replace(',',''))
df_budget['worldwide_gross'] = df_budget['worldwide_gross'].astype(float)
df_budget['production_budget'] = df_budget.production_budget.apply(lambda x:x.replace('$','').replace(',',''))

df_budget['production_budget'] = df_budget['production_budget'].astype(float)


df_budget
```

```python
#It's also valuable to identify in which seasons there is an overall good worldwide gross value
season_dict = {'Jan': 'Winter',
               'Feb': 'Winter',
               'Mar': 'Spring', 
               'Apr': 'Spring',
               'May': 'Spring',
               'Jun': 'Summer',
               'Jul': 'Summer',
               'Aug': 'Summer',
               'Sep': 'Fall',
               'Oct': 'Fall',
               'Nov': 'Fall',
               'Dec': 'Winter'}
df_budget['Season'] = df_budget['Month'].apply(lambda x: season_dict[x])
```

```python
#we only let the data that is more updated
recent_movie_years = df_budget.query("Year >= 2012")
recent_movie_years
```

```python
fig = plt.subplots( figsize=(16, 5))

films_years = recent_movie_years['Year'].value_counts()
sns.lineplot(x=films_years.index, y=films_years.values, data=films_years)
plt.xlabel('Year')
plt.ylabel('Number of Films Released')
plt.show()
```

```python
top_months = recent_movie_years.groupby('Month')[['worldwide_gross']].mean().sort_values(by='worldwide_gross', ascending=False)
top_months
```

```python
# Now we can reset the index so that the first column is called "Months"

top_months.index.name = "Month"
top_months.reset_index(inplace=True)
top_months
```

```python
fig = plt.subplots( figsize=(16, 5))

months = top_months['Month']
gross_profit = top_months['worldwide_gross']
# Plot data
plt.bar(months, gross_profit)

# Customize appearance
plt.tick_params(axis="x", labelrotation=30)
plt.title('Monthly Gross Revenue')
plt.xlabel('Month of Release')
plt.ylabel("Average Gross Revenue (in 100 Millions)")
```

```python
df_budget
```

```python
seasons_gross = recent_movie_years.groupby('Season')[['worldwide_gross']].mean().sort_values(by='worldwide_gross', ascending=False)
seasons_gross
```

```python
seasons_gross.index.name = "Season"
seasons_gross.reset_index(inplace=True)
seasons_gross
```

```python
fig = plt.subplots( figsize=(16, 3))

season = seasons_gross['Season']
gross = seasons_gross['worldwide_gross']
# Plot data
plt.bar(season, gross)

# Customize appearance
plt.tick_params(axis="x", labelrotation=30)
plt.title('Seasonal Gross Revenue')
plt.xlabel('Season')
plt.ylabel("Average Gross Revenue (in 100 Millions)")

#Summer is the season where films have higher average gross revenue
```

```python
graph = sns.FacetGrid(df_budget, col ="Season",  hue ="Season")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "production_budget", "worldwide_gross", edgecolor ="w").add_legend()
# show the object
plt.show()


#we observe that we have a few outliers, but there is a positive tendency as film studios invest more, 
# their worldwide gross value tend to be higher
```
